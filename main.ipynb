```python
temp1=ratings.merge(books,on='ISBN')
```

```python
temp1
```

```python
ratings_with_name=temp1
```

```python
def convert_to_numeric(value):
    try:
        return float(value)
    except ValueError:
        return 0
```

```python
ratings_with_name['Book-Rating'] = ratings_with_name['Book-Rating'].apply(convert_to_numeric)
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df.rename(columns={'Book-Rating':'num-ratings'},inplace=True)
```

```python
num_rating_df
```

```python
ratings_with_name['Book-Rating'].fillna(0, inplace=True)
```

```python
ratings_with_name
```

```python
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce').fillna(0).astype(int)
```

```python
ratings_with_name
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean(numeric_only=True).reset_index()
```

```python
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
```
