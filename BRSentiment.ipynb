```python
from textblob import TextBlob
```

```python
def calculate_sentiment(text):
    # Create a TextBlob object with the user's sentiment text
    blob = TextBlob(str(text))
    # Calculate the polarity score (ranging from -1 to 1)
    sentiment_score = blob.sentiment.polarity
    return sentiment_score
```

```python
users['Sentiment_Score'] = users['Sentiment'].apply(calculate_sentiment)
```

```python
merged_df = pd.merge(pd.merge(users, rating, on='User-ID'), books, on='ISBN')
```
